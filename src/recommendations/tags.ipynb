```python
import pandas as pd
import string
import re
import json
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import torch
from transformers import AutoTokenizer, AutoModel
```

```python
tags = ['ds/ml/dl', 'front, back, fullstack', 'design', 'ctf',
'product/project manager', 'analysts and marketers', 'gamedev', 'robotics', "искусственный интеллект", "машинное обучение", "дизайн", "разработка"]
test = "Самый масштабный проект по искусственному интеллекту в России"
```

```python
stop_words = stopwords.words("russian")
tokens = word_tokenize(test, language="russian")
filtered_tokens = []
for token in tokens:
    if token not in stop_words:
      if token not in [".", ",", "/", "?", "!", ";", ":"]:
        filtered_tokens.append(token)
```

```python
filtered_tokens
```

```python
tokenizer = AutoTokenizer.from_pretrained("cointegrated/LaBSE-en-ru")
model = AutoModel.from_pretrained("cointegrated/LaBSE-en-ru")
```

```python
encoded_input0 = tokenizer(tags, padding=True, truncation=True, max_length=64, return_tensors='pt')
encoded_input1 = tokenizer(filtered_tokens, padding=True, truncation=True, max_length=8, return_tensors='pt')
with torch.no_grad():
    model_output0 = model(**encoded_input0)
    model_output1 = model(**encoded_input1)
embeddings0 = model_output0.pooler_output
embeddings0 = torch.nn.functional.normalize(embeddings0).numpy()
embeddings1 = model_output1.pooler_output
embeddings1 = torch.nn.functional.normalize(embeddings1).numpy()
top_n = 5
distances = cosine_similarity(embeddings0, embeddings1)
a = distances.argsort()[0][-top_n:]
keywords = [tags[index] for index in distances.argsort()[0][-top_n:]]
keywords
```
